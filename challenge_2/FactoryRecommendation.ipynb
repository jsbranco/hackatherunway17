```python
import pandas as pd

tool1matches = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool1.xlsm', sheetname = 'Matches')
tool1data = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool1.xlsm', sheetname = 'Data')
tool2matches = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool2.xlsm', sheetname = 'Matches')
tool2data = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool2.xlsm', sheetname = 'Data')
tool3matches = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool3.xlsm', sheetname = 'Matches')
tool3data = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool3.xlsm', sheetname = 'Data')
tool4matches = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool4.xlsm',sheetname = 'Matches')
tool4data = pd.read_excel('C://Users/Dhruv/Downloads/Challenge - Matching Tool/tool4.xlsm',sheetname = 'Data')
```

```python
# Stripping column names of white spaces

tool1data = tool1data.rename(columns = lambda x : x.strip())
tool2data = tool1data.rename(columns = lambda x : x.strip())
tool3data = tool1data.rename(columns = lambda x : x.strip())
tool4data = tool1data.rename(columns = lambda x : x.strip())

tool1data['Vendor_No'] = tool1data['Vendor_No'].str.replace('Vendor ','')
tool2data['Vendor_No'] = tool2data['Vendor_No'].str.replace('Vendor ','')
tool3data['Vendor_No'] = tool3data['Vendor_No'].str.replace('Vendor ','')
tool4data['Vendor_No'] = tool4data['Vendor_No'].str.replace('Vendor ','')
```

```python
frames = [tool1data,tool2data,tool3data,tool4data]
tools = pd.concat(frames)
tools.head()
```

```python
matchFrames = [tool1matches,tool2matches,tool3matches,tool4matches]
matches = pd.concat(matchFrames)
matches.head()
```

```python
yesDict = {'DV':1, 'Y':1,'N':0}
```

```python
matches['MatchValue'] = matches['MatchValue'].map(yesDict)

only_matches = matches[matches['MatchValue']==1]
only_matches = only_matches[['MatchValue','Unique.No.x','Unique.No.y']]
only_matches.head()
```

```python
dataMatches = pd.merge(tools, only_matches, how='left', left_on='Unique.No', right_on='Unique.No.x')
dataMatches.shape
```

```python
dataMatches['Year'] = dataMatches['Year'].map(lambda x : x.strip())
dataMatches['Year'] = dataMatches['Year'].map(lambda x : x.split(' '))
dataMatches['Year'] = dataMatches['Year'].map(lambda x : x[1])

dataMatches['Season_Name'] = dataMatches['Season_Name'].map(lambda x : x.strip())
dataMatches['Season_Name'] = dataMatches['Season_Name'].map(lambda x : x.split(' '))
dataMatches['Season_Name'] = dataMatches['Season_Name'].map(lambda x : x[0])

def vartoCat(colNames):
    for col in colNames:
        dataMatches[col] = dataMatches[col].astype('category')
        dataMatches[col] = dataMatches[col].cat.codes
    return dataMatches

colstoCat = ['TypeProduct_Desc','Retail_Class_Desc','Retail_SubClass_Desc','Season_Name',
             'Gender_Desc','Size_Cd','Size_Cd','Country_Cd','Silhouette_Cd']
vartoCat(colstoCat)
dataMatches.head()
```

```python
import contextlib
import itertools
from math import sqrt
from operator import add
import sys

from docopt import docopt
from pyspark import SparkConf, SparkContext
from pyspark.mllib.recommendation import ALS
```

```python
SPARK_EXECUTOR_MEMORY = '4g'
SPARK_APP_NAME = 'MatchingChallenge'
SPARK_MASTER = 'local'


@contextlib.contextmanager
def spark_manager():
    conf = SparkConf().setMaster(SPARK_MASTER) \
                      .setAppName(SPARK_APP_NAME) \
                      .set("spark.executor.memory", SPARK_EXECUTOR_MEMORY)
    spark_context = SparkContext(conf=conf)

    try:
        yield spark_context
    finally:
        spark_context.stop()
```
